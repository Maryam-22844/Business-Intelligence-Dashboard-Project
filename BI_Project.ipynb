Handling Missing Values

```python
df.isna().sum()
```

```python
# ID column --> Giving it a unique ID
distinct_id = max(df['ID']) + 1
df['ID'].fillna(distinct_id, inplace = True)
df["ID"] = df["ID"].astype(str)
```

```python
# AMOUNT, PRICE, LINENET, and LINENETTOTAL columns -- replacing by mean
df['AMOUNT'].fillna(df['AMOUNT'].mean(), inplace = True)
df['PRICE'].fillna(df['PRICE'].mean(), inplace = True)
df['LINENET'].fillna(df['LINENET'].mean(), inplace = True)
df['LINENETTOTAL'].fillna(df['AMOUNT'].mean(), inplace = True)
```

```python
# DATE_ -- using forward fill method
df['DATE_'] = df['DATE_'].fillna(method='pad')
```

```python
# CITY column -- There is only missing value and we dropped it
missing_city_rows = df[df['CITY'].isnull()]
print(missing_city_rows)
df = df.dropna(subset=['CITY'])
```

```python
df.isna().sum()
```

Using the Dendogram Hierarchies ! filling the missing values

```python
from sklearn.impute import SimpleImputer
subset_df = df[['CLIENTCODE', 'CLIENTNAME', 'GENDER']]

# Perform imputation using SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
subset_filled = pd.DataFrame(imputer.fit_transform(subset_df), columns=['CLIENTCODE', 'CLIENTNAME', 'GENDER'])

# Replace the missing values in the original DataFrame with the imputed values
df[['CLIENTCODE', 'CLIENTNAME', 'GENDER']] = subset_filled[['CLIENTCODE', 'CLIENTNAME', 'GENDER']]
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
subset_df = df[['ITEMCODE', 'ITEMNAME', 'CATEGORY_NAME1', 'CATEGORY_NAME2']]

# Perform imputation using SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
subset_filled = pd.DataFrame(imputer.fit_transform(subset_df), columns=['ITEMCODE', 'ITEMNAME', 'CATEGORY_NAME1', 'CATEGORY_NAME2'])

# Replace the missing values in the original DataFrame with the imputed values
df[['ITEMCODE', 'ITEMNAME', 'CATEGORY_NAME1', 'CATEGORY_NAME2']] = subset_filled[['ITEMCODE', 'ITEMNAME', 'CATEGORY_NAME1', 'CATEGORY_NAME2']]
```

```python
from sklearn.impute import SimpleImputer
subset_df = df[['BRANDCODE', 'BRAND']]

# Perform imputation using SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
subset_filled = pd.DataFrame(imputer.fit_transform(subset_df), columns=['BRANDCODE', 'BRAND'])

# Replace the missing values in the original DataFrame with the imputed values
df[['BRANDCODE', 'BRAND']] = subset_filled[['BRANDCODE', 'BRAND']]
```

Outliere in the LINENET Column (Main KPI)

```python
# Display descriptive statistics for LINENET
print(df_Copied['LINENET'].describe())

# Calculate IQR (Interquartile Range)
Q1 = df_Copied['LINENET'].quantile(0.25)
Q3 = df_Copied['LINENET'].quantile(0.75)
IQR = Q3 - Q1

# Identify potential outliers using IQR method
outliers = (df_Copied['LINENET'] < (Q1 - 1.5 * IQR)) | (df_Copied['LINENET'] > (Q3 + 1.5 * IQR))
print("Potential Outliers:")
print(df_Copied[outliers]['LINENET'])
```

```python
# Create a boxplot for LINENET
plt.boxplot(df_Copied['LINENET'])
plt.xlabel('LINENET')
plt.title('Boxplot of LINENET')
plt.show()
```

Finding Kurtosis and Skew -- To deal with Outliers

```python
from scipy.stats import skew, kurtosis
# Calculate skewness and kurtosis
skewness = skew(df_Copied['LINENET'])
kurt = kurtosis(df_Copied['LINENET'])

print("Skewness:", skewness)
print("Kurtosis:", kurt)
```

Before Dealing Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import probplot

# Plot the histogram
plt.figure(figsize=(12, 6))
sns.histplot(df_Copied['LINENET'], bins=50, kde=True, color='blue', label='LINENET Distribution')
plt.xlabel('LINENET')
plt.ylabel('Frequency')
plt.title('Histogram of LINENET')
plt.legend()
plt.show()
```

```python
percentile25 = df_Copied['LINENET'].quantile(0.25)
percentile75 = df_Copied['LINENET'].quantile(0.75)
iqr = percentile75 - percentile25
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
upper_limit
```

```python
lower_limit
```

```python
df_Copied[df_Copied["LINENET"] > upper_limit]
```

```python
df_Copied[df_Copied["LINENET"] < lower_limit]
```

```python
outlier_mask = (df_Copied['LINENET'] > upper_limit) | (df_Copied['LINENET'] < lower_limit)
outlier_rows = df_Copied[outlier_mask]
df_Copied['outlier'] = 'N'
df_Copied.loc[outlier_mask, 'outlier'] = 'Y'
```

```python
import xlwings as xw
# Create a new Excel workbook
wb = xw.Book()
sheet = wb.sheets.active
# Write the DataFrame to the Excel sheet starting from cell A1
sheet.range('A1').value = df_Copied
wb.save('cleaned_with_outliers.xlsx')
wb.close()
```

Capping to handle  LINENET outliers

```python
import numpy as np
new_df_cap = df_Copied.copy()

new_df_cap['LINENET'] = np.where(
    new_df_cap['LINENET'] > upper_limit, upper_limit,
    np.where(new_df_cap['LINENET'] < lower_limit, lower_limit,
             new_df_cap['LINENET']
            )
)
```

Anova Test check the relationship between GENDER and LINENET

```python
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('LINENET ~ C(Q("GENDER"))', data=new_df_cap).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print ("\nAnova => LINENET - GENDER")
display(anova_table)
```

The p-value (0.13375) is greater than the typical significance level of 0.05, suggesting that there is no significant difference in 'LINENET' among different gender categories. In other words, based on the given analysis, it does not appear that 'GENDER' significantly explains the variance in 'LINENET'.

Anova Test check the relationship between BRAND and LINENET

```python
model = ols('LINENET ~ C(Q("BRAND"))', data=new_df_cap).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print ("\nAnova => LINENET - BRAND")
display(anova_table)
```

The p-value for the 'BRAND' variable is very low (close to 0.0), indicating that there is a significant effect of 'BRAND' on 'LINENET.'
Since the p-value is less than the conventional significance level (e.g., 0.05), you would reject the null hypothesis and conclude that there are significant differences in the mean 'LINENET' values across different levels of 'BRAND.'
The F-statistic of 60.63 further supports the evidence of a significant effect.
